```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("training_data.csv")

steps = df["step"]
means = df["mean_reward"]
stds  = df["std_reward"]

# Calculate and print average return with standard deviation for last 100,000 steps
last_100k = df[df["step"] >= df["step"].max() - 100000]
avg_return = last_100k["mean_reward"].mean()
std_return = last_100k["mean_reward"].std()
print(f"PPO - Average Return (last 100k steps): {avg_return:.2f} Â± {std_return:.2f}")

# Plot main line
plt.plot(steps, means, color="blue", label="Mean Reward")

# Transparent blue error band
plt.fill_between(
    steps,
    means - stds,
    means + stds,
    color="blue",
    alpha=0.2,     # transparency (0 = invisible, 1 = opaque)
    label="Std Dev"
)

plt.xlabel("Training Step")
plt.ylabel("Reward")
plt.title("PPO training reward")
plt.legend()
plt.show()
```

```python
# Load all three datasets
df_ppo = pd.read_csv("training_data.csv")
df_td3 = pd.read_csv("training_data_TD3.csv")
df_sac = pd.read_csv("training_data_SAC.csv")

# Create figure
plt.figure(figsize=(12, 6))

# PPO
steps_ppo = df_ppo["step"]
means_ppo = df_ppo["mean_reward"]
stds_ppo = df_ppo["std_reward"]
plt.plot(steps_ppo, means_ppo, color="blue", label="PPO", linewidth=2)
plt.fill_between(steps_ppo, means_ppo - stds_ppo, means_ppo + stds_ppo, color="blue", alpha=0.15)

# TD3
steps_td3 = df_td3["step"]
means_td3 = df_td3["mean_reward"]
stds_td3 = df_td3["std_reward"]
plt.plot(steps_td3, means_td3, color="red", label="TD3", linewidth=2)
plt.fill_between(steps_td3, means_td3 - stds_td3, means_td3 + stds_td3, color="red", alpha=0.15)

# SAC
steps_sac = df_sac["step"]
means_sac = df_sac["mean_reward"]
stds_sac = df_sac["std_reward"]
plt.plot(steps_sac, means_sac, color="green", label="SAC", linewidth=2)
plt.fill_between(steps_sac, means_sac - stds_sac, means_sac + stds_sac, color="green", alpha=0.15)

# Calculate and print average returns for last 100k steps
last_100k_ppo = df_ppo[df_ppo["step"] >= df_ppo["step"].max() - 100000]
avg_ppo = last_100k_ppo["mean_reward"].mean()
std_ppo = last_100k_ppo["mean_reward"].std()

last_100k_td3 = df_td3[df_td3["step"] >= df_td3["step"].max() - 100000]
avg_td3 = last_100k_td3["mean_reward"].mean()
std_td3 = last_100k_td3["mean_reward"].std()

last_100k_sac = df_sac[df_sac["step"] >= df_sac["step"].max() - 100000]
avg_sac = last_100k_sac["mean_reward"].mean()
std_sac = last_100k_sac["mean_reward"].std()

print(f"PPO - Average Return (last 100k steps): {avg_ppo:.2f} Â± {std_ppo:.2f}")
print(f"TD3 - Average Return (last 100k steps): {avg_td3:.2f} Â± {std_td3:.2f}")
print(f"SAC - Average Return (last 100k steps): {avg_sac:.2f} Â± {std_sac:.2f}")

plt.xlabel("Training Step")
plt.ylabel("Reward")
plt.title("Training Reward Comparison: PPO vs TD3 vs SAC")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
